## Problem 1: Running Sum of 1D Array

You are given an array of integers `nums`.

The running sum is defined as:
runningSum[i] = sum of elements from index 0 to i.

Return the running sum of the array.


```python
class Solution:
    def runningSum(self, nums):
        result = []
        current_sum = 0

        for num in nums:
            current_sum += num
            result.append(current_sum)

        return result
```

```python
obj = Solution()
print(obj.runningSum([1, 2, 3, 4]))
```

## Problem 2: Shuffle the Array

You are given an array `nums` of length `2n`.

The first `n` elements belong to the first half
and the next `n` elements belong to the second half.

Rearrange and return the array as:
[x1, y1, x2, y2, ..., xn, yn]


```python
class Solution:
    def shuffle(self, nums, n):
        result = []

        for i in range(n):
            result.append(nums[i])
            result.append(nums[i + n])

        return result
```

```python
obj = Solution()
print(obj.shuffle([2,5,1,3,4,7], 3))
```

## Problem 3: Kids With the Greatest Number of Candies

You are given an array `candies` and an integer `extraCandies`.

For each child, check if giving all extraCandies makes their total candies
greater than or equal to the maximum candies any child has.

Return a list of True or False.


```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        max_candies = max(candies)
        result = []

        for candy in candies:
            result.append(candy + extraCandies >= max_candies)

        return result
```

```python
obj = Solution()
print(obj.kidsWithCandies([2,3,5,1,3], 3))
```
