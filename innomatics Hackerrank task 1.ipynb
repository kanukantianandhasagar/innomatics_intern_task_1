### HackerRank – Python If-Else

You are given an integer `n`.

Print:
- `Weird` if `n` is odd  
- `Not Weird` if `n` is even and between 2 and 5  
- `Weird` if `n` is even and between 6 and 20  
- `Not Weird` if `n` is even and greater than 20


```python


import math
import os
import random
import re
import sys

if __name__ == '__main__':
    n = int(input().strip())

    if n % 2 != 0:
        print("Weird")
    elif 2 <= n <= 5:
        print("Not Weird")
    elif 6 <= n <= 20:
        print("Weird")
    else:
        print("Not Weird")
```

5.## Problem: Python Loops
